## **Introduction**
Machine learning had supported the agriculture in numerous ways. It had helped in predicting yeild, quality, efficiency and descison making. In this notebook, we try to analyse the Harvest season dataset and evaluate different models in prediction whether the crop would be health, damaged by pesticides or damaged by other reaons.

## **Data Overview**
The data provided consists of almost 4600 observations
  * **ID** : Unique ID
  * **Estimated_Insects_Count** : Estimated insects count per square meter
  * **Crop_Type** : Category of crop (0 or 1)
  * **Soil_Type** : Category of soil (0 or 1)
  * **Pesticide_Use_Category** : Types of pesticides uses (1-*Never* , 2-*Previously Used*, 3-*Currently Using*)
  * **Number_Doses_Week** : Number of doses per week
  * **Number_Weeks_Used** : Number of weeks used
  * **Number_Weeks_Quit** : Number of weeks quit
  * **Season** : season category (1,2, or 3)
  * **Crop_Damage** : Crop damage category (0-*Alive*, 1-*Damage due to other causes*, 2*-Damage due to pesticides*)

## **Data Preparation**

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

* There are null values in the Number weeks used feature

```python
# Check for duplicates
data.duplicated().sum()
```

```python
# Check data description
data.describe().T
```

```python
# Count values in each columns
for i in data.columns:
  print(i)
  print(data[i].nunique())
  print(data[i].value_counts(),end='\n\n')
```

* The ID has all unique values and it doesnt provide any significant information, therefore we can drop it.
* The labels in the target variable is highly imbalance, we need to apply oversampling technique before training the models

```python
# Drop ID column
data.drop(columns=['ID'],inplace=True)
```

```python
data.groupby('Crop_Type')['Number_Weeks_Used'].mean()
```

```python
data.groupby('Soil_Type')['Number_Weeks_Used'].mean()
```

```python
data.groupby('Season')['Number_Weeks_Used'].mean()
```

* The mean value of number weeks used is almost same for individual Crop types and by season
* The mean of number_weeks_used is different for groups in soil type, therefore we can use the mean of soil groups to fill up the nulls values present.

```python
# Fill the null values
data.loc[(data['Number_Weeks_Used'].isnull()) & (data['Soil_Type'] == 0),'Number_Weeks_Used'] = data.groupby('Soil_Type')['Number_Weeks_Used'].mean()[0]
data.loc[(data['Number_Weeks_Used'].isnull()) & (data['Soil_Type'] == 1),'Number_Weeks_Used'] = data.groupby('Soil_Type')['Number_Weeks_Used'].mean()[1]
```

```python
# Check for null values
data.isnull().sum()
```

```python
data.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid", {'grid.linestyle': '--'})
sns.set_context('notebook')
```

```python
# Define a function for calculating total outliers and extreme outliers
def calc_outliers(x,ret=False):
  a = x.quantile(0.25)
  b = x.quantile(0.75)
  IQR = b-a

  if not ret:
    print('Total Outliers : ',round((((x < (a-IQR*1.5)) | (x > (b+IQR*1.5))).sum()/len(x)).iloc[0]*100,2),'%')
    print('Extreme Outliers : ',round((((x < (a-IQR*3)) | (x > (b+IQR*3))).sum()/len(x)).iloc[0]*100,2),'%')
  else:
    return ((x < (a-IQR*3)) | (x > (b+IQR*3)))
```

## **Feature Engineering**

```python
data.columns
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()
x_smote, y_smote = smote.fit_resample(x, y)

print('Original dataset shape')
print(y.value_counts())
print('Resample dataset shape')
print(round(pd.DataFrame(y_smote)).value_counts())
```

```python
x = round(pd.DataFrame(x_smote,columns=x.columns))
y = round(pd.DataFrame(y_smote,columns=['Crop_Damage']))
```

```python
enc = []
for i in ['Crop_Type','Soil_Type','Pesticide_Use_Category','Season']:
    enc.append(pd.get_dummies(x[i],prefix=i,drop_first=True))

encoded = pd.concat(enc,axis=1)
encoded.shape
```

```python
x.drop(columns=['Crop_Type','Soil_Type','Pesticide_Use_Category','Season'],inplace=True)
```

```python
x = pd.concat([x,encoded],axis=1)
```

```python
x.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaled = MinMaxScaler().fit_transform(x)
```

```python
x = pd.DataFrame(scaled,columns=x.columns)
```

## **Model Evaluation and Training**

```python
from sklearn.model_selection import train_test_split, KFold, GridSearchCV, StratifiedKFold, cross_val_score,cross_validate
from sklearn.linear_model import RidgeClassifier, LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, ExtraTreesClassifier
```

```python
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score, precision_score,recall_score,roc_curve, roc_auc_score, classification_report
```

```python
# Find out random state for best possible arrangement of train and test data
random_state = []
for i in range(1,101):
  x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=i)
  model = LogisticRegression().fit(x_train,y_train)
  random_state.append(model.score(x_test,y_test))
print('Best random state : ',random_state.index(max(random_state))+1)
```

```python
# Split data into train and test data
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=86)
```

```python
# Load linear, tree and ensemble model
models_dict = {
    'LGC': LogisticRegression(),
    'RDC': RidgeClassifier(),
    'LDA': LinearDiscriminantAnalysis(),
    'KNC': KNeighborsClassifier(),
    'DTC': DecisionTreeClassifier(),
    'RFC': RandomForestClassifier(),
    'ABC': AdaBoostClassifier(),
    'ETC': ExtraTreesClassifier(),
}
```

```python
# Apply cross validation on each models
cv_score = {}
for model in models_dict:
  print('-----',model,'-----')
  cls = models_dict[model]
  kfold = StratifiedKFold(n_splits=9,random_state=49)
  cv_results = cross_validate(cls, x_train, y_train, scoring='accuracy', cv=kfold)
  cv_score[model] = cv_results['test_score']
  print('R2 Score : ',round(cv_results['test_score'].mean(),3),' +/-',round(cv_results['test_score'].std()*2,3))
```

```python
# Plot the accuracies of the classfiers
df = pd.DataFrame(cv_score)
plt.figure(figsize=(10,5))
sns.boxplot(data=df)
plt.title('Algorithms vs Accuracy')
plt.show()
```

* Knn, Decision tree, random forest and extra tree classifiera were performing better. Lets do parameter tuning for further imrpovement

### **Hyper parameter tuning**

```python
# KNN classifier
params = {
    'leaf_size': list(range(1,6)),
    'n_neighbors': list(range(1,10)),
    'p': [1,2],
    }

kfold = KFold(n_splits=9,random_state=49)
knc_grid = GridSearchCV(KNeighborsClassifier(), param_grid=params, scoring='accuracy', cv=kfold)
knc_grid.fit(x_train,y_train)
print('Best: %f using %s' % (knc_grid.best_score_, knc_grid.best_params_))
```

```python
# Random Forest Classifier
params = {
    'n_estimators': [100, 150],
    'max_depth': [3,4,5,6],
    'max_features': ["sqrt"],
    'min_samples_split': [2,4,8],
    'bootstrap': [True, False]
    }

kfold = KFold(n_splits=9,random_state=49)
rfc_grid = GridSearchCV(RandomForestClassifier(), param_grid=params, scoring='accuracy', cv=kfold)
rfc_grid.fit(x_train,y_train)
print('Best: %f using %s' % (rfc_grid.best_score_, rfc_grid.best_params_))
```

```python
# Extra Trees Classifier
params = {
    'n_estimators': [50,75,100,125,150],
    'criterion':['gini','entropy'],
    'min_samples_leaf': [1,2,3,4,5],
    'min_samples_split': [2,3,4],
    }

kfold = KFold(n_splits=9,random_state=49)
etc_grid = GridSearchCV(ExtraTreesClassifier(), param_grid=params, scoring='accuracy', cv=kfold)
etc_grid.fit(x_train,y_train)
print('Best: %f using %s' % (etc_grid.best_score_, etc_grid.best_params_))
```

```python
# DTC classifier
params = {
    'criterion':['gini', 'entropy'],
    'max_depth':[2,4,6,8,10,12]
    }

kfold = KFold(n_splits=9,random_state=49)
dtc_grid = GridSearchCV(DecisionTreeClassifier(), param_grid=params, scoring='accuracy', cv=kfold)
dtc_grid.fit(x_train,y_train)
print('Best: %f using %s' % (dtc_grid.best_score_, dtc_grid.best_params_))
```

### **Predictions**

```python
models = [dtc_grid,knc_grid,rfc_grid,etc_grid]
model_names = ['LGR','KNN','RFC','ETC']
for i in range(len(models)):
  pred = models[i].predict(x_test)
  print(model_names[i])
  print(classification_report(y_test,pred))
```

```python
sns.heatmap(confusion_matrix(y_test,etc_grid.best_estimator_.predict(x_test)),annot=True)
plt.show()
```
