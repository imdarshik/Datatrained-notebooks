## **Introduction**
In this notebook we try to analyse the US covid reporting from april 12th to December 28th and try to evaluate reggression algorithms on polynomial features by trying to predict the deaths and recovered cases. In this notebook, I have also attempted to train model for forecasting deaths and predictions.

## **Data Overview**

  * **Province_State** - The name of the State within the USA.
  * **Country_Region** - The name of the Country (US).
  * **Last_Update** - The most recent date the file was pushed.
  * **Lat** - Latitude.
  * **Long_** - Longitude.
  * **Confirmed** - Aggregated confirmed case count for the state.
  * **Deaths** - Aggregated Death case count for the state.
  * **Recovered** - Aggregated Recovered case count for the state.
  * **Active** - Aggregated confirmed cases that have not been resolved (Active =   * Confirmed - Recovered - Deaths).
  * **FIPS** - Federal Information Processing Standards code that uniquely identifies counties within the USA.
  * **Incident_Rate** - confirmed cases per 100,000 persons.
  * **People_Tested** - Total number of people who have been tested.
  * **People_Hospitalized** - Total number of people hospitalized.
  * **Mortality_Rate** - Number recorded deaths * 100/ Number confirmed cases.
  * **UID** - Unique Identifier for each row entry.
  * **ISO3** - Officialy assigned country code identifiers.
  * **Testing_Rate** - Total number of people tested per 100,000 persons.
  * **Hospitalization_Rate** - Total number of people hospitalized * 100/ Number of confirmed cases.

* In the row 6609, the neagtive value in the Active feature is due to the abnormal counting in the recovered cases of Minnesota. So lets smooth the value using the mean of previous and next observations

* The latitude and longitude of Diamond princess and grand princess are null, these are the names of the ships. Which doesnt have any stationary cordinates. We can drop these rows, since we require only the state data for the analysis.
* There exists lot of null values in the recovered case feature, which we can fill it by taking the difference of confirmed cases and death and active cases.

```python
# Drop rows with names of cruise ship
data = data[data['Province_State']!='Diamond Princess']
data = data[data['Province_State']!='Grand Princess']
```

```python
# Fill the null values of recovered column
data.loc[(data['Recovered'].isnull()),'Recovered'] = (data[data['Recovered'].isnull()].Confirmed - 
                                    (data[data['Recovered'].isnull()].Active + 
                                    data[data['Recovered'].isnull()].Deaths))
```

```python
# check for null avlues
dc = {}
for i in data.Province_State.unique():
  dc[i] = dict(data[data['Province_State'] == i].isnull().sum())
pd.DataFrame(dc).iloc[:,1:-1].T
```

## **Feature Engineering**

```python
data.columns
```

```python
df = data.groupby('Date')[['Confirmed','Active','Deaths','Recovered']].sum()
```

```python
df['Date'] = np.arange(1,len(df)+1)
```

```python
x = df[['Confirmed','Active','Date']]
y_deaths = df[['Deaths']]
y_recovered = df[['Recovered']]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
x = MinMaxScaler().fit_transform(x)
```

## **Model Training and Evaluation**

```python
from sklearn.model_selection import train_test_split,KFold, cross_val_score, GridSearchCV,cross_validate
```

```python
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline, make_pipeline
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Create a dictionary for pipelines
models_dict = {
    'LNR': Pipeline([('LNR', LinearRegression())]),
    'L2R': Pipeline([('L2',PolynomialFeatures(degree=2)),('L2R', LinearRegression())]),
    'L3R': Pipeline([('L3',PolynomialFeatures(degree=3)),('L3R', LinearRegression())]),
    'L4R': Pipeline([('L4',PolynomialFeatures(degree=4)),('L4R', LinearRegression())]),
    'RDG': Pipeline([('RDG', Ridge())]),
    'R2R': Pipeline([('R2',PolynomialFeatures(degree=2)),('R2R', Ridge())]),
    'R3R': Pipeline([('R3',PolynomialFeatures(degree=3)),('R3R', Ridge())]),
    'R4R': Pipeline([('R4',PolynomialFeatures(degree=4)),('R4R', Ridge())]),
}
```

### **Predicting deaths**

```python
x_train, x_test, y_train, y_test = train_test_split(x,y_deaths,test_size=0.2,shuffle=False)
```

```python
# Evaluate models
predictions = {}
score = {}
for i in models_dict:
  print('----------',i,'----------')
  model = models_dict[i].fit(x_train,y_train)
  pred = model.predict(x_test)
  print('MAE : ',mean_absolute_error(y_test,pred))
  print('MSE : ',mean_squared_error(y_test,pred))
  print('RMSE : ',np.sqrt(mean_squared_error(y_test,pred)))
  print('R2Score : ',r2_score(y_test,pred))
  print('TrScore : ',model.score(x_train,y_train))
  predictions[i] = pred
  score[i] = [r2_score(y_test,pred)]
```

* The linear regression model had achieved 99% training score whereas the r2 score is negative. This mean the model is overfitted
* The Ridge classifier with polynomial features of degree 3 was able to achieve better performace with 86% r2 score and 96% training score.

```python
model = models_dict['R3R'].fit(x_train,y_train)
pred = model.predict(x_test)
```

```python
plt.title('True vs Prediction')
plt.plot(np.arange(len(x_test)),y_test,label='True')
plt.plot(np.arange(len(x_test)),pred,label='prediction')
plt.legend()
plt.show()
```

```python
import joblib
joblib.dump(model,'r3r_estimator_deaths.pkl')
```

## **Death forecasting**

```python
df_death = data.groupby('Date',as_index=False)[['Date','Deaths']].sum()
```

```python
# Find the difference of each rows
df_data = df_death[['Deaths']].diff()
df_data.index = df_death['Date']
```

```python
df_data.dropna(inplace=True)
```

```python
# plot the data along the date
plt.figure(figsize=(15,3))
sns.lineplot(data=df_data,x='Date',y='Deaths')
plt.show()
```

* The has fluctuations and doesnt have any notable trend

```python
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
```

```python
# Use seasonal decompose to find out the trend and seasonality
seasonal_decompose = sm.tsa.seasonal_decompose(df_data, model='additive',extrapolate_trend='freq')
fig = seasonal_decompose.plot()
fig.set_size_inches(15,5)
plt.show()
```

* Here we dont see any notable trend.
* There is a weekly seasonality in the data

```python
# Find moving average
rolmean = df_data.rolling(window=7).mean() 
rolstd = df_data.rolling(window=7).std()

plt.figure(figsize=(15,3))
plt.plot(df_data, label='Original')
plt.plot(rolmean, label='Rolling Mean')
plt.plot(rolstd, label='Rolling Std')
plt.legend()
plt.show()
```

* The rolling mean and rolling standard deviation are not stationary, we need to apply detrending or differencing method to stationarize the data

```python
# Apply detrending
df_data_detrend = (df_data - df_data.rolling(window=7).mean()) / df_data.rolling(window=7).std()
```

```python
rolmean = df_data_detrend.rolling(window=7).mean() 
rolstd = df_data_detrend.rolling(window=7).std()

plt.figure(figsize=(15,3))
plt.plot(df_data_detrend, label='Original')
plt.plot(rolmean, label='Rolling Mean')
plt.plot(rolstd, label='Rolling Std')
plt.legend()
plt.show()
```

* The moving average and standard deviations appear to be stationary
* Lets do augmented dickey fuller test to confirm the data is stationary

```python
adftest = adfuller(df_data_detrend.dropna()['Deaths'], autolag='AIC')
print('Test Statistic : ',round(adftest[0],3))
print('P-Value : ',round(adftest[1],3))
print('Critical Values :')
for t, ts in adftest[4].items():
  print(t,' : ',ts)
```

* The test statitic is very lower than the 5% critical value. Hence the data is non stationary

```python
# Apply differencing and find out the data will be stationary
df_data_diff = df_data - df_data.shift(7)
```

```python
rolmean = df_data_diff.rolling(window=7).mean() 
rolstd = df_data_diff.rolling(window=7).std()

plt.figure(figsize=(15,3))
plt.plot(df_data_diff, label='Original')
plt.plot(rolmean, label='Rolling Mean')
plt.plot(rolstd, label='Rolling Std')
plt.legend()
plt.show()
```

* Appears to be stationary

```python
df_data_diff.dropna(inplace=True)
adftest = adfuller(df_data_diff.dropna()['Deaths'], autolag='AIC')
print('Test Statistic : ',round(adftest[0],3))
print('P-Value : ',round(adftest[1],3))
print('Critical Values :')
for t, ts in adftest[4].items():
  print(t,' : ',ts)
```

* After applying the ADF test we see that the the test statistc is still lower. Hence the data is non stationary

```python
# Combining both detrending and differencing
df_data_dtrend_diff = df_data_detrend - df_data_detrend.shift(7)
```

```python
rolmean = df_data_dtrend_diff.rolling(window=7).mean() 
rolstd = df_data_dtrend_diff.rolling(window=7).std()

plt.figure(figsize=(15,3))
plt.plot(df_data_dtrend_diff, label='Original')
plt.plot(rolmean, label='Rolling Mean')
plt.plot(rolstd, label='Rolling Std')
plt.legend()
plt.show()
```

```python
df_data_dtrend_diff.dropna(inplace=True)
adftest = adfuller(df_data_dtrend_diff.dropna()['Deaths'], autolag='AIC')
print('Test Statistic : ',round(adftest[0],3))
print('P-Value : ',round(adftest[1],3))
print('Critical Values :')
for t, ts in adftest[4].items():
  print(t,' : ',ts)
```

* By combining detrending and differencing, the test statistic has become higher. Now the data is stationary with 99% confidence.

### **Training model**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Split data into 80% for training and 20% for validation
train = df_data[:208]
val = df_data[208:]
predict_dates = len(df_data) - len(train)
print(train.shape)
print(val.shape)
print(predict_dates)
```

```python
from statsmodels.tsa.api import ExponentialSmoothing
```

* Since the data has seasonality we will use Holt-Winters seasonal method

```python
model = ExponentialSmoothing(train,seasonal_periods=7,trend='add',seasonal='add',freq="D").fit(use_boxcox='log')
pred = model.forecast(predict_dates)
```

```python
plt.figure(figsize=(15,3))
plt.plot(val, label='True')
plt.plot(pred, label='Forecasted')
plt.legend()
plt.show()
```

```python
print('MAE',mean_absolute_error(val,pred))
print('MSE',mean_squared_error(val,pred))
print('RMSE',np.sqrt(mean_squared_error(val,pred)))
print('R2Score',r2_score(val,pred))
```

* So the model has achieved an RMSE of 583 and R2score of 52%

```python
import joblib
joblib.dump(model,'ExponentialSmoothing_forecast_covid.pkl')
```
