```python
freq_quals = ['mbbs','md','bds','mds','bams','bhms','ms','dnb','dlo','fellowship','ddvl','pgdd','ddv','dhms','phd']
```

```python
from fuzzywuzzy import process

def match_names(wrong_names,correct_names):
    names_array=[]
    for row in wrong_names:
        x=process.extractOne(row, correct_names)
        if x[1]<60:
            names_array.append('Others')
        else:
            names_array.append(x[0])
    return names_array
```

```python
train_data.loc[:,'Qualification'] = match_names(
    train_data.Qualification.astype(str),
    freq_quals)
```

```python
train_qual = train_data.Qualification.agg(lambda x: pd.Series(1,(set(x.strip().lower().split(','))))).fillna(0)
train_qual.reset_index(drop=True,inplace=True)
train_qual.head()
```

```python
test_data.loc[:,'Qualification'] = match_names(
    test_data.Qualification.astype(str),
    freq_quals)
```

```python
df1 = train_data.Place.agg(lambda x: ' '.join(re.findall('[a-z]+',''.join(str(x).lower().strip().split(',')[:-1])))).rename('Area')
df2 = train_data.Place.agg(lambda x: ' '.join(re.findall('[a-z]+',''.join(str(x).lower().strip().split(',')[-1]))))

df = pd.concat([df1,df2,train_data.Fees],axis=1,)
df_mean = df.groupby(['Place','Area']).mean()

area_dict = {}

for i in ['mumbai','hyderabad','chennai','delhi','bangalore','coimbatore','thiruvananthapuram','ernakulam']:
    area_dict[i] = dict(zip(list(df_mean.loc[i].index),list(df_mean.loc[i].Fees.values)))
    df.loc[df.Place==i,'Area'] = df.loc[df.Place==i].Area.replace(area_dict[i])
    
df.Area = df.Area.astype(float)
df_area = df.Area
```
